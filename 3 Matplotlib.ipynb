```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,20,50,60]
```

```python
#scatter plot
plt.scatter(x, y)
```

```python
#to display values
fig,ax = plt.subplots()

for i,j in zip(x,y):
    ax.annotate('(%s,' %i, xy=(i,j))
    ax.annotate('%s)' %j, xy=(i,j), xytext=(25,0), textcoords='offset points')
    
    
plt.scatter(x, y)
```

```python
#to draw line as well
x = [1,2,3,4,5,6]
y = [10,20,30,20,50,60]

fig,ax = plt.subplots()

for i,j in zip(x,y):
    ax.annotate('(%s,' %i, xy=(i,j))
    ax.annotate('%s)' %j, xy=(i,j), xytext=(25,0), textcoords='offset points')
    
    
plt.scatter(x, y)
plt.plot(x,y , linewidth=1)
```

```python
#Sine wave:
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
#t starts from 0, 0.01,0.02.......1.99
s = np.sin(2*np.pi * t)

plt.plot(t, s)
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
```

```python
#Subplots:
```

```python
x = np.arange(0.0, 5.0, 0.1)

y1 = np.sin(2 * np.pi * x) * np.exp(-x)
y2 = np.sin(2 * np.pi * x)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.ylabel('Damped oscillation')
plt.title('Subplots:')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.ylabel('Undamped oscillation')


```

```python
x = np.arange(0.0, 5.0, 0.1)

y1 = np.sin(2 * np.pi * x) * np.exp(-x)
y2 = np.sin(2 * np.pi * x)

y3 = np.tan(2 * np.pi * x)  * np.exp(-x)
y4 = np.tan(2 * np.pi * x)

plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'o-')
plt.ylabel('Sin Damped')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, '.-')
plt.ylabel('Sin')

plt.subplot(2, 2, 3)
plt.plot(x1, y3, 'o-')
plt.ylabel('tan Damped')

plt.subplot(2, 2, 4)
plt.plot(x2, y4, '.-')
plt.ylabel('tan')
```

```python
#Bar chart:
```

```python
country = ['China','India','US','Indonesia','Brazil']
poplation = [23, 15 , 10 , 8 , 6 ]

index = np.arange(len(country))
plt.bar(index, poplation, align='center', alpha=0.5)
plt.xticks(index, country)
```

```python
#Horizontal Bar chart:
country = ['China','India','US','Indonesia','Brazil']
poplation = [23, 15 , 10 , 8 , 6 ]

index = np.arange(len(country))
plt.barh(index, poplation, align='center', alpha=0.5)
plt.yticks(index, country)
```

```python
#Stacked bar chart:
```

```python
# data to plot

country = ['China','India','US','Indonesia','Brazil']
poplationMale = [13, 8 , 10 , 9 , 9 ]
poplationFemale = [15, 5 , 12 , 8 , 6 ]

 
index = np.arange(len(country))
bar_width = 0.35
 
plt.bar(index, poplationMale, bar_width,
                 alpha=0.8,
                 color='b',
                 label='Male')
 
plt.bar(index + bar_width, poplationFemale, bar_width,
                 alpha=0.8,
                 color='g',
                 label='Female')
 
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(index + bar_width , country)
plt.legend()
```

```python
#pie chart
country = ['China','India','US','Indonesia','Brazil']
poplation = [23, 15 , 10 , 8 , 6 ]

plt.pie(poplation,  labels=country, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal') #to make it look like a circle
plt.show()
```

```python
#Donut chart:

#draw a circle at the center of pie to make it look like a donut

country = ['China','India','US','Indonesia','Brazil']
poplation = [23, 15 , 10 , 8 , 6 ]

plt.pie(poplation, labels=country, 
        autopct='%1.1f%%', shadow=True)


#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.7,color='black', fc='white',linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```
